# GitHub Personal Access Token Auto Approve Policy Modified

Detects when changes are made to a GitHub organization's fine-grained Personal Access Token approval policy.

## Detection Query
Execute the following query to find alerts related to this rule.

```python
%ingest.source_type:github
action:auto_approve_personal_access_token_requests.disable OR action:auto_approve_personal_access_token_requests.enable OR action:forbid_auto_approve_personal_access_token_requests.disable OR action:forbid_auto_approve_personal_access_token_requests.enable
```

## Investigation Steps
1. **Identify the Actor**: Who performed the action?
2. **Review User Activity**: What else did this user do around the same time?
3. **Check IP Reputation**: Is the source IP known to be malicious?
4. **Verify Authorization**: Was this action planned or approved?
