# GitHub SSO Configuration Modified

Detects when changes are made to a GitHub enterprise or organization's Single Sign-On (SSO) configuration.

## Detection Query
Execute the following query to find alerts related to this rule.

```python
%ingest.source_type:github
action:business.disable_oidc OR action:business.disable_saml OR action:business.update_saml_provider_settings OR action:org.disable_saml OR action:org.update_saml_provider_settings OR action:org_credential_authorization.deauthorize OR action:org_credential_authorization.grant OR action:org_credential_authorization.revoke OR action:sso_redirect.disable
```

## Investigation Steps
1. **Identify the Actor**: Who performed the action?
2. **Review User Activity**: What else did this user do around the same time?
3. **Check IP Reputation**: Is the source IP known to be malicious?
4. **Verify Authorization**: Was this action planned or approved?
