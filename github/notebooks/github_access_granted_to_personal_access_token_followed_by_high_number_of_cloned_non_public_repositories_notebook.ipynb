# GitHub Access Granted To Personal Access Token Followed By High Number Of Cloned Non Public Repositories

Detects when a user grants access to a GitHub Personal Access Token prior to cloning several GitHub non-public GitHub repositories. An adversary may grant access to a Personal Access Token before attempting to steal the contents of several GitHub repositories using a an automated script or offensive tool.

## Detection Query
Execute the following query to find alerts related to this rule.

```python
%ingest.source_type:github
action:git.clone OR action:personal_access_token.access_granted
```

## Investigation Steps
1. **Identify the Actor**: Who performed the action?
2. **Review User Activity**: What else did this user do around the same time?
3. **Check IP Reputation**: Is the source IP known to be malicious?
4. **Verify Authorization**: Was this action planned or approved?

```python
%ingest.source_type:github
actor: <ACTOR_FROM_ALERT>
| count(action) by action
```
