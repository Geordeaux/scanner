# GitHub Secret Scanning Disabled Or Bypassed

Detects when components of GitHub secret scanning is disabled for an enterprise or organization.

## Detection Query
Execute the following query to find alerts related to this rule.

```python
%ingest.source_type:github
action:business_secret_scanning.disable OR action:business_secret_scanning.disabled_for_new_repos OR action:business_secret_scanning_custom_pattern_push_protection.disabled OR action:business_secret_scanning_push_protection.disable OR action:business_secret_scanning_push_protection.disabled_for_new_repos OR action:business_secret_scanning_push_protection_custom_message.disable OR action:org_secret_scanning_generic_secrets.disabled OR action:repository_secret_scanning.disable OR action:repository_secret_scanning_generic_secrets.disabled OR action:repository_secret_scanning_push_protection.disable OR action:secret_scanning.disable OR action:secret_scanning_new_repos.disable OR action:secret_scanning_push_protection.bypass
```

## Investigation Steps
1. **Identify the Actor**: Who performed the action?
2. **Review User Activity**: What else did this user do around the same time?
3. **Check IP Reputation**: Is the source IP known to be malicious?
4. **Verify Authorization**: Was this action planned or approved?
