# GitHub organization was transferred between enterprise accounts

Detect when a GitHub organization is transferred between enterprise accounts.

## Detection Query
Execute the following query to find alerts related to this rule.

```python
%ingest.source_type:github
action:org.transfer
| groupbycount(actor)
| where @q.count > 0
```

## Investigation Steps
1. **Identify the Actor**: Who performed the action?
2. **Review User Activity**: What else did this user do around the same time?
3. **Check IP Reputation**: Is the source IP known to be malicious?
4. **Verify Authorization**: Was this action planned or approved?
