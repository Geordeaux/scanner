# GitHub Enterprise Or Organization Recovery Codes Activity

Detects when the Single Sign-On (SSO) recovery codes are accessed for a GitHub enterprise or organization. An adversary may attempt to access or steal SSO recovery codes in order to maintain access to their target's GitHub environment.

## Detection Query
Execute the following query to find alerts related to this rule.

```python
%ingest.source_type:github
action:business.recovery_codes_downloaded OR action:business.recovery_codes_generated OR action:business.recovery_codes_printed OR action:business.recovery_codes_viewed OR action:org.recovery_codes_downloaded OR action:org.recovery_codes_generated OR action:org.recovery_codes_printed OR action:org.recovery_codes_viewed
```

## Investigation Steps
1. **Identify the Actor**: Who performed the action?
2. **Review User Activity**: What else did this user do around the same time?
3. **Check IP Reputation**: Is the source IP known to be malicious?
4. **Verify Authorization**: Was this action planned or approved?
