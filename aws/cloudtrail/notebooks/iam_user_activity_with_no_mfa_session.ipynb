# Investigation: IAM User Activity with No MFA Session

This notebook assists in investigating this detection.

```python
from scanner_client import ScannerClient
import pandas as pd
import plotly.express as px

# Initialize Scanner Client
client = ScannerClient(api_url="https://api.kast-apse1.scanner.dev", api_key="YOUR_API_KEY")
```

```python
query = """
%ingest.source_type:aws:cloudtrail
userIdentity.type:"IAMUser"
userIdentity.sessionContext.attributes.mfaAuthenticated:"false"
NOT userAgent:("cloudformation.amazonaws.com" OR "application-autoscaling.amazonaws.com" OR "AWS Internal")
"""

results = client.query(query, time_range_s=86400)
df = pd.DataFrame(results)
df.head()
```
