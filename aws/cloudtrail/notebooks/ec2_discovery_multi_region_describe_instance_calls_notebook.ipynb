# High Frequency of EC2 Multi-Region DescribeInstances API Calls

Identifies when a user makes multiple `DescribeInstances` API calls in multiple regions within a short window.
High frequency of `DescribeInstances` API calls across multiple regions may indicate an adversary attempting to discover the EC2 instances in the account or perform reconnaissance.

## Detection Query
Execute the following query to find alerts related to this rule.

```python
%ingest.source_type:aws:cloudtrail
eventName:DescribeInstances
eventSource:"ec2.amazonaws.com"
| stats count_distinct(awsRegion) as region_count by userIdentity.arn
| where region_count >= 10
```

## Investigation Steps
1. **Identify the Principal**: Who performed the action?
2. **Review Context**: What else did this principal do?
3. **Check Permissions**: Did the principal have legitimate reasons for this action?
4. **Verify Resource**: Is the resource critical or sensitive?
