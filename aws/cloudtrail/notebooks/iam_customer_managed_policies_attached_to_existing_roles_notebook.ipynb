# AWS IAM Customer-Managed Policy Attachment to Existing Roles

Identifies when customer-managed IAM policies are attached to existing roles.
This could indicate a privilege escalation attempt by attaching policies with elevated permissions.

## Detection Query
Execute the following query to find alerts related to this rule.

```python
%ingest.source_type:aws:cloudtrail
eventName:AttachRolePolicy
eventSource:"iam.amazonaws.com"
| parse_json requestParameters "policyArn"
| where NOT policyArn matches "^arn:aws:iam::aws:policy"
```

## Investigation Steps
1. **Identify the Principal**: Who performed the action?
2. **Review Context**: What else did this principal do?
3. **Check Permissions**: Did the principal have legitimate reasons for this action?
4. **Verify Resource**: Is the resource critical or sensitive?
