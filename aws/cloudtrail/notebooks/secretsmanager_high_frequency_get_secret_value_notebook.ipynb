# Secrets Manager High Frequency GetSecretValue API Calls

Identifies high frequency of `GetSecretValue` API calls to AWS Secrets Manager programmatically.
This may indicate an adversary attempting to access sensitive information.

## Detection Query
Execute the following query to find alerts related to this rule.

```python
%ingest.source_type="aws:cloudtrail"
eventName="GetSecretValue"
eventSource="secretsmanager.amazonaws.com"
NOT userAgent:("Chrome" OR "Firefox" OR "Safari" OR "Edge" OR "Brave" OR "Opera")
| stats count() as request_counts by userIdentity.arn
| where request_counts >= 10
```

## Investigation Steps
1. **Identify the Principal**: Who performed the action?
2. **Review Context**: What else did this principal do?
3. **Check Permissions**: Did the principal have legitimate reasons for this action?
4. **Verify Resource**: Is the resource critical or sensitive?
