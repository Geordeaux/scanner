# SES Identity Has Been Deleted

Detects an instance of an SES identity being deleted via the "DeleteIdentity" event. This may be an indicator of an adversary removing the account that carried out suspicious or malicious activities

References:
* https://unit42.paloaltonetworks.com/compromised-cloud-compute-credentials/

False Positives:
* Unknown

## Detection Query
Execute the following query to find alerts related to this rule.

```python
%ingest.source_type="aws:cloudtrail"
eventSource="ses.amazonaws.com"
eventName="DeleteIdentity"
| stats
min(@scnr.datetime) as firstTime,
max(@scnr.datetime) as lastTime,
count() as eventCount
by
userIdentity.arn,
eventSource,
eventName,
awsRegion
```

## Investigation Steps
1. **Identify the Principal**: Who performed the action?
2. **Review Context**: What else did this principal do?
3. **Check Permissions**: Did the principal have legitimate reasons for this action?
4. **Verify Resource**: Is the resource critical or sensitive?
