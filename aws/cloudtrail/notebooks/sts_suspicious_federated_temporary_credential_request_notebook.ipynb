# STS Suspicious Federated Temporary Credential Request

This hunting query identifies when a user requests temporary federated credentials with a duration greater than 24 hours or with the `AdministratorAccess` policy attached. Federated users are typically given temporary credentials to access AWS services. A duration greater than 24 hours or the `AdministratorAccess` policy attached may indicate an adversary attempting to maintain access to AWS services for an extended period of time or escalate privileges.

## Detection Query
Execute the following query to find alerts related to this rule.

```python
%ingest.source_type="aws:cloudtrail"
eventSource="sts.amazonaws.com"
eventName="GetFederationToken"
(requestParameters like "*AdministratorAccess*" OR requestParameters like "*durationSeconds*")
| where requestParameters.durationSeconds > 86400 OR requestParameters like "*AdministratorAccess*"
```

## Investigation Steps
1. **Identify the Principal**: Who performed the action?
2. **Review Context**: What else did this principal do?
3. **Check Permissions**: Did the principal have legitimate reasons for this action?
4. **Verify Resource**: Is the resource critical or sensitive?
