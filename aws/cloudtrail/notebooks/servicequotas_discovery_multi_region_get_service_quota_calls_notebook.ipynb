# High Frequency of Service Quotas Multi-Region GetServiceQuota API Calls

Identifies when a user makes `GetServiceQuota` API calls for EC2 on-demand instances in multiple regions.
This could indicate a threat actor attempting to discover quota limits for cryptomining.

## Detection Query
Execute the following query to find alerts related to this rule.

```python
%ingest.source_type="aws:cloudtrail"
eventSource="servicequotas.amazonaws.com"
eventName="GetServiceQuota"
requestParameters:("ec2" AND "L-1216C47A")
| stats countdistinct(awsRegion) as region_count by userIdentity.arn
| where region_count >= 10
```

## Investigation Steps
1. **Identify the Principal**: Who performed the action?
2. **Review Context**: What else did this principal do?
3. **Check Permissions**: Did the principal have legitimate reasons for this action?
4. **Verify Resource**: Is the resource critical or sensitive?
