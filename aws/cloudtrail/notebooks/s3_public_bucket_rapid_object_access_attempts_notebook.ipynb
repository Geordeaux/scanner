# S3 Public Bucket Rapid Object Access Attempts

Identifies when an anonymous user makes multiple `GetObject` requests to a public S3 bucket.
Rapid access may indicate data exfiltration or reconnaissance.

## Detection Query
Execute the following query to find alerts related to this rule.

```python
%ingest.source_type="aws:cloudtrail"
eventName="GetObject"
eventSource="s3.amazonaws.com"
userIdentity.type="Anonymous"
| stats count() as access_count by requestParameters.bucketName
| where access_count >= 15
```

## Investigation Steps
1. **Identify the Principal**: Who performed the action?
2. **Review Context**: What else did this principal do?
3. **Check Permissions**: Did the principal have legitimate reasons for this action?
4. **Verify Resource**: Is the resource critical or sensitive?
