# An AWS S3 bucket lifecycle expiration policy was set to disabled

Detect if an AWS S3 lifecycle expiration policy is disabled in your CloudTrail logs.

## Detection Query
Execute the following query to find alerts related to this rule.

```python
(not errorCode:*)
eventSource:s3.amazonaws.com
eventName:PutBucketLifecycle
requestParameters.LifecycleConfiguration.Rule.Expiration:*
requestParameters.LifecycleConfiguration.Rule.Status:Disabled
| groupbycount(requestParameters.bucketName)
| where @q.count > 0
```

1. **Identify the Principal**: Who performed the action?
2. **Review Context**: What else did this principal do?
3. **Check Permissions**: Did the principal have legitimate reasons for this action?
4. **Verify Resource**: Is the resource critical or sensitive?
