# High EC2 Instance Deployment Count Attempts

Identifies when a user makes EC2 `RunInstances` API calls with a high instance deployment count.
High instance deployment counts may indicate an adversary attempting to deploy a large number of instances for cryptomining or other malicious activities.

## Detection Query
Execute the following query to find alerts related to this rule.

```python
%ingest.source_type:aws:cloudtrail
eventName:RunInstances
eventSource:"ec2.amazonaws.com"
| parse_json requestParameters "minCount", "maxCount"
| eval target_instance_count = maxCount - minCount + 1
| where target_instance_count >= 10
```

## Investigation Steps
1. **Identify the Principal**: Who performed the action?
2. **Review Context**: What else did this principal do?
3. **Check Permissions**: Did the principal have legitimate reasons for this action?
4. **Verify Resource**: Is the resource critical or sensitive?
