# AWS principal added to multiple EKS clusters

Detect when an AWS principal is granted permissions on multiple Amazon EKS clusters in a short period, indicating a potential compromise of your AWS environment.

## Detection Query
Execute the following query to find alerts related to this rule.

```python
%ingest.source_type:"aws:cloudtrail"
eventName:(CreateAccessEntry AssociateAccessPolicy)
| stats countdistinct(requestParameters.name) as num_distinct_values by requestParameters.principalArn, recipientAccountId
| where num_distinct_values > 5
```

## Investigation Steps
1. **Identify the Principal**: Who performed the action?
2. **Review Context**: What else did this principal do?
3. **Check Permissions**: Did the principal have legitimate reasons for this action?
4. **Verify Resource**: Is the resource critical or sensitive?
