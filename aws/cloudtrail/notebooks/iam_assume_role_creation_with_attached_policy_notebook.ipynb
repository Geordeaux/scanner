# IAM Assume Role Creation with Attached Policy

Identifies the creation of an IAM role with an inline attached policy that allows the role to assume another role in a different account.
This could indicate a potential backdoor IAM role.

## Detection Query
Execute the following query to find alerts related to this rule.

```python
%ingest.source_type:aws:cloudtrail
eventName:CreateRole
eventSource:"iam.amazonaws.com"
requestParameters:("AssumeRole" AND "allow" AND "arn:aws:iam")
| regex requestParameters "AWS\": \"arn:aws:iam::(?<target_account_id>\d+):"
| where target_account_id != recipientAccountId
```

## Investigation Steps
1. **Identify the Principal**: Who performed the action?
2. **Review Context**: What else did this principal do?
3. **Check Permissions**: Did the principal have legitimate reasons for this action?
4. **Verify Resource**: Is the resource critical or sensitive?
