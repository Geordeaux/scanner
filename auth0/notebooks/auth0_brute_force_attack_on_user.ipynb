# Investigation: Auth0 Brute Force Attack

This notebook assists in investigating brute force attacks against users.

```python
from scanner_client import ScannerClient
import pandas as pd
import plotly.express as px

# Initialize Scanner Client
client = ScannerClient(api_url="https://api.kast-apse1.scanner.dev", api_key="YOUR_API_KEY")
```

```python
query = """
%ingest.source_type:auth0
data.type:(fp fu)
"""

results = client.query(query, time_range_s=3600)
df = pd.DataFrame(results)

if not df.empty:
    ip_counts = df['data.ip'].value_counts().reset_index()
    ip_counts.columns = ['IP Address', 'Count']
    fig = px.bar(ip_counts, x='IP Address', y='Count', title='Failed Logins by IP')
    fig.show()
```
