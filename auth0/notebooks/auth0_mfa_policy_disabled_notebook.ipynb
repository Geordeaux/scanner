# Auth0 MFA Policy Disabled

Detect when MFA policies are disabled in Auth0. This can weaken
security controls and should be monitored closely.

## Detection Query
Execute the following query to find alerts related to this rule.

```python
%ingest.source_type:auth0
data.description:'Set the Multi-factor Authentication policies'
data.details.request.path:/api/v2/guardian/policies
data.details.request.body:"[]"
data.details.request.channel:'https://manage.auth0.com'
| groupbycount(data.user_id)
| where @q.count > 0
```

## Investigation Steps
1. **Identify the Actor**: Who performed the action?
2. **Review Context**: What else did this user do?
3. **Check Location**: Was the action performed from a usual location?
4. **Verify Intent**: Was this action business-justified?
