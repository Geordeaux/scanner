# Auth0 New Admin Invited FOLLOWED BY Tenant Member Account Deletion

A user was invited as admin and shortly after deleted tenant member accounts. This may indicate account takeover attempts.

## Detection Query
Execute the following query to find alerts related to this rule.

```python
%ingest.source_type:auth0
type:s OR type:f
```

## Investigation Steps
1. **Identify the Actor**: Who performed the action?
2. **Review Context**: What else did this user do?
3. **Check Location**: Was the action performed from a usual location?
4. **Verify Intent**: Was this action business-justified?

```python
%ingest.source_type:auth0
data.user_name: <USER_NAME>
| count(type) by type
```
