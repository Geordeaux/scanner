# Investigation: Auth0 Credential Stuffing

This notebook assists in investigating credential stuffing attacks.

```python
from scanner_client import ScannerClient
import pandas as pd
import plotly.express as px

# Initialize Scanner Client
client = ScannerClient(api_url="https://api.kast-apse1.scanner.dev", api_key="YOUR_API_KEY")
```

```python
query = """
%ingest.source_type:auth0
data.type:(scoa fcoa pwd_leak)
"""

results = client.query(query, time_range_s=3600)
df = pd.DataFrame(results)

if not df.empty:
    type_counts = df['data.type'].value_counts().reset_index()
    type_counts.columns = ['Event Type', 'Count']
    fig = px.bar(type_counts, x='Event Type', y='Count', title='Credential Stuffing Events')
    fig.show()
```
