# Auth0 User Deleted

Detects when a user is deleted in Auth0. While this can be a legitimate administrative
action, unauthorized deletion of users can disrupt operations or be used to cover
tracks after a compromise.

## Detection Query
Execute the following query to find alerts related to this rule.

```python
%ingest.source_type:auth0
data.type:sapi
data.description:"Delete a user"
data.details.request.path:/api/v2/users/*
| groupbycount(data.user_id)
| where @q.count > 0
```

## Investigation Steps
1. **Identify the Actor**: Who performed the action?
2. **Review Context**: What else did this user do?
3. **Check Location**: Was the action performed from a usual location?
4. **Verify Intent**: Was this action business-justified?
