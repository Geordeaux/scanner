# Auth0 suspicious IP throttling disabled

Detect when suspicious IP throttling is disabled in Auth0.

## Detection Query
Execute the following query to find alerts related to this rule.

```python
%ingest.source_type:auth0
data.type:sapi
data.details.request.path:/api/v2/attack-protection/suspicious-ip-throttling
data.details.response.body.enabled:false
| groupbycount(data.client_id)
| where @q.count > 0
```

## Investigation Steps
1. **Identify the Actor**: Who performed the action?
2. **Review Context**: What else did this user do?
3. **Check Location**: Was the action performed from a usual location?
4. **Verify Intent**: Was this action business-justified?
