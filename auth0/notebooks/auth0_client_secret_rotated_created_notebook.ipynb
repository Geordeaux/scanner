# Auth0 Client Secret Rotated or Created

Detects when a client secret is rotated or created in Auth0. Attackers may
rotate secrets to maintain persistence or gain unauthorized access to
applications.

## Detection Query
Execute the following query to find alerts related to this rule.

```python
%ingest.source_type:auth0
data.type:sapi
(
(data.description:"Rotate a client secret") or
(data.description:"Create a client secret")
)
data.details.request.path:/api/v2/clients/*/rotate-secret
| groupbycount(data.user_id)
| where @q.count > 0
```

## Investigation Steps
1. **Identify the Actor**: Who performed the action?
2. **Review Context**: What else did this user do?
3. **Check Location**: Was the action performed from a usual location?
4. **Verify Intent**: Was this action business-justified?
