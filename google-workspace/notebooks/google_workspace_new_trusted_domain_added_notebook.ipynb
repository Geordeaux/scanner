# Google Workspace New Trusted Domain Added

Identifies when a domain is added to the list of trusted domains in Google Workspace. An adversary may attempt to manipulate sharing settings for trusted domains to gain unauthorized access to sensitive files and folders within an organization.

## Detection Query
Execute the following query to find alerts related to this rule.

```python
%ingest.source_type:google_workspace
action:ADD_TRUSTED_DOMAINS
```

## Investigation Steps
1. **Identify the User**: Who performed the action?
2. **Review Context**: What else did this user do?
3. **Check Location**: Was the action performed from a usual location?
4. **Verify Intent**: Was this action business-justified?
