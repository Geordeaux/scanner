# Google Workspace MFA Disabled

Identifies when multi-factor authentication (MFA) is disabled for a Google Workspace organization. Security teams can monitor for changes to MFA configuration that may weaken the organization's security posture.

## Detection Query
Execute the following query to find alerts related to this rule.

```python
%ingest.source_type:google_workspace
action:ALLOW_STRONG_AUTHENTICATION OR action:ENFORCE_STRONG_AUTHENTICATION
```

## Investigation Steps
1. **Identify the User**: Who performed the action?
2. **Review Context**: What else did this user do?
3. **Check Location**: Was the action performed from a usual location?
4. **Verify Intent**: Was this action business-justified?
