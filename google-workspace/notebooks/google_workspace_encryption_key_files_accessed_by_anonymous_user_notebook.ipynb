# Google Workspace Encryption Key File Accessed By An Anonymous User

Identifies when an encryption key file is accessed by an anonymous user in Google Drive. An adversary may attempt to access encryption keys before moving laterally and compromising sensitive data.

## Detection Query
Execute the following query to find alerts related to this rule.

```python
%ingest.source_type:google_workspace
action:copy OR action:download OR action:view
```

## Investigation Steps
1. **Identify the User**: Who performed the action?
2. **Review Context**: What else did this user do?
3. **Check Location**: Was the action performed from a usual location?
4. **Verify Intent**: Was this action business-justified?

```python
%ingest.source_type:google_workspace
actor.email: <USER_EMAIL>
| count(action) by action
```
