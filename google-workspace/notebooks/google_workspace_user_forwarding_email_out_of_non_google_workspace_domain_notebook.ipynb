# Google Workspace user forwarding email out of non Google Workspace domain

Create a signal when Google Workspace detects a user setting up mail forwarding to a non-Google Workspace domain.

## Detection Query
Execute the following query to find alerts related to this rule.

```python
%ingest.source_type:gsuite
event.name:email_forwarding_out_of_domain
| groupbycount(actor.email)
| where @q.count > 0
```

## Investigation Steps
1. **Identify the User**: Who performed the action?
2. **Review Context**: What else did this user do?
3. **Check Location**: Was the action performed from a usual location?
4. **Verify Intent**: Was this action business-justified?
