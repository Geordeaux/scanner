# Google Workspace Suspicious Login and Google Drive File Share

Identifies when a Google Workspace user shares a file on Google Drive after a suspicious login event occurred.

## Detection Query
Execute the following query to find alerts related to this rule.

```python
%ingest.source_type:google_workspace
action:change_acl_editors OR action:change_document_access_scope OR action:change_document_visibility OR action:change_user_access OR action:login_success
```

## Investigation Steps
1. **Identify the User**: Who performed the action?
2. **Review Context**: What else did this user do?
3. **Check Location**: Was the action performed from a usual location?
4. **Verify Intent**: Was this action business-justified?
