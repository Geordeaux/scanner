# Google Cloud SQL database modified

Detect when a Google Cloud SQL database has been modified.

## Detection Query
Execute the following query to find alerts related to this rule.

```python
%ingest.source_type:gcp
protoPayload.methodName:(cloudsql.instances.create cloudsql.instances.create cloudsql.users.create cloudsql.users.update)
| groupbycount(project_id, database_id, protoPayload.authenticationInfo.principalEmail)
| where @q.count > 0
```

## Investigation Steps
1. **Identify the Principal**: Who performed the action?
2. **Review Context**: What else did this principal do?
3. **Check Permissions**: Did the principal have legitimate reasons for this action?
4. **Verify Resource**: Is the resource critical or sensitive?
