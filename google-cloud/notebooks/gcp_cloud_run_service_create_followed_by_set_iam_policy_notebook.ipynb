# GCP Cloud Run Service Created FOLLOWED BY Set IAM Policy

Detects run.services.create method for privilege escalation in GCP. The exploit creates a new Cloud Run Service that, when invoked, returns the Service Account's access token by accessing the metadata API of the server it is running on.

## Detection Query
Execute the following query to find alerts related to this rule.

```python
%ingest.source_type:google_cloud_audit
protoPayload.methodName:"google.cloud.run.v1.Services.CreateService" OR protoPayload.methodName:"google.iam.v1.IAMPolicy.SetIamPolicy"
```

## Investigation Steps
1. **Identify the Principal**: Who performed the action?
2. **Review Context**: What else did this principal do?
3. **Check Permissions**: Did the principal have legitimate reasons for this action?
4. **Verify Resource**: Is the resource critical or sensitive?
