# Google Cloud Project external principal added as project owner

Detect attempts to add an external principal as a project owner in Google Cloud.

## Detection Query
Execute the following query to find alerts related to this rule.

```python
%ingest.source_type:gcp
protoPayload.serviceName:cloudresourcemanager.googleapis.com
protoPayload.methodName:InsertProjectOwnershipInvite
data.protoPayload.request.member:(*gmail.com *googlemail.com)
| groupbycount(project_id, protoPayload.authenticationInfo.principalEmail)
| where @q.count > 0
```

## Investigation Steps
1. **Identify the Principal**: Who performed the action?
2. **Review Context**: What else did this principal do?
3. **Check Permissions**: Did the principal have legitimate reasons for this action?
4. **Verify Resource**: Is the resource critical or sensitive?
