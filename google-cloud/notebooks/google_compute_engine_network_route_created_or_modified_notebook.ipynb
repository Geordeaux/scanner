# Google Compute Engine network route created or modified

Detect when a firewall rule is created or modified.

## Detection Query
Execute the following query to find alerts related to this rule.

```python
%ingest.source_type:gcp
protoPayload.methodName:(beta.compute.routes.insert beta.compute.routes.patch)
| groupbycount(project_id, data.protoPayload.resourceName, protoPayload.authenticationInfo.principalEmail)
| where @q.count > 0
```

## Investigation Steps
1. **Identify the Principal**: Who performed the action?
2. **Review Context**: What else did this principal do?
3. **Check Permissions**: Did the principal have legitimate reasons for this action?
4. **Verify Resource**: Is the resource critical or sensitive?
